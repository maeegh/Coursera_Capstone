<h2>Introduction</h2>

Throughout history, bread has been a staple food. At its most basic, bread is simply flour and water, mixed and heated. Increasingly complex recipes were developed over time, introducing processes such as leavening, layering, and even adding flavours. Bread became so integral to many peoples that it is now considered a fundamental aspect of culinary, social, and religious culture.

<h3>Business Problem</h3>

György has just moved to Budapest after 40 years in Székelyföld (more commonly known as Transylvania), and he is keen to bring his culinary talents to the people of Budapest. He is immensely proud of his heritage, and believes the people of Budapest will enjoy his bread.
However, despite having a good Unique Selling Proposition, György would like to open his bakery in one of the less crowded markets. Although he is not destitute, György is not a wealthy man, and does not wish to take on any more risk than he has to. Attempting to open a bakery in an area where there are already many bakeries is too risky for him. György is fairly mobile, and is willing to open his bakery in any part of Budapest - just so long as it's not too high competition.

This project will analyse the districts of Budapest, and through several methodologies and techniques learned in this IBM Data Science course, help György to know where he can open his new bakery, so he can provide his neighbours with some wonderful Székely treats.

<h3>Background</h3>

![Szekely Land](https://upload.wikimedia.org/wikipedia/commons/5/55/T%C3%B6rt_Sz%C3%A9kelyf%C3%B6ld.png)

The Székelys are ethnic Hungarians, living in what is now Romania. Despite centuries of identification with the Magyar people, the Treaty of Trianon in 1920 forcibly separated the Székely from Hungary, isolating them in increasingly smaller parts of Romania. The Székely enjoy a long cultural tradition of literature and food, which have been maintained regardless of political land disputes.

One of the most well-known breads is Potato Battered Bread, a light bread with a good crust, perfect to accompany heavier meals such as gulyás (Goulash) or paprikás csirke (Paprika Chicken) on cold winter evenings. Potatoes, yeast, four and water come together to create a soft and tender dough that's baked over 90 minutes to produce one of Székelyland's best exports.

<h3>Target Audience</h3>

Although this project has been created to directly answer György's query, it can be equally useful to anyone wishing to open a bakery in Budapest. Hungarians like their bread, so although bakeries may be plentiful, there is almsot always room for one more - as long as it has sufficient market potential.
Furthermore, anyone with a niche bakery product cannot operate a bakery solely on this one product. Having a good UVP is not sufficient - Hungarians will expect typical bakery products to be sold alongside. Therefore, it is necessary to consider the overall vitality of a bakery, rather than focussing on one singular product.

<h3>Data Used</h3>

* List of Neighbourhoods in Budapest
* Venue data
* Specific location (latitude and longitude)

<h3>Data Source</h3>

* Wikipedia - https://en.wikipedia.org/wiki/Category:Neighbourhoods_of_Budapest, which is scraped using the beautifulsoup library
* Foursquare - provides venue information for neighbourhoods. This allows us to use the neighbourhood data and then create a new dataframe specifically for bakeries.
* Clustering/k-means - machine learning allows us to place the results into clusters, giving György both a general region, and specific locales in which he might consider opening his bakery. 
* Folium - allows for the visualisation of data
* Geocoder - provides latitude and longitudinal data
